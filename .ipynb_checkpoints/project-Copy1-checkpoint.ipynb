```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt  
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('FF_allwaves_2019.csv')
```

```python
data.columns
data = data.drop(['Unnamed: 0'], axis=1)
```

```python
data.columns
```

```python
data[['m5k2a','m5k2b','m5k2c','m5k2d', 'ch5ppvtraw']]
```

```python
m9_agg = data[['m5k2a','m5k2b','m5k2c','m5k2d','ch5ppvtraw']]
m9_agg = m9_agg.rename(columns={"m5k2a": "hard", "m5k2b": "trapped", "m5k2c":"work", "m5k2d": "tired",'ch5ppvtraw':'ppvt_raw'})
# filter missing/not in wave
m9_agg = m9_agg[(m9_agg['work']!=-6)&(m9_agg['work']!=-9)&(m9_agg['work']!=-3)]
col_list = m9_agg.columns
```

```python
# summing up aggravation
col_list = ['hard','trapped','work','tired']
m9_agg['aggravation'] = m9_agg[col_list].sum(axis=1)
# standardizing aggravation per person
#m9_agg['std'] = m9_agg.std(axis= 1)
#m9_agg['mean'] = m9_agg[col_list].mean(axis=1)
m9_agg
```

```python
# fill NaN for ppvt not in wave
m9_agg.loc[m9_agg['ppvt_raw'] == -9,'ppvt_raw'] = np.nan
```

```python
sns.regplot('aggravation', 'ppvt_raw', data = m9_agg).set_title('Mother Aggravation vs PPVT')
```

```python
results = smf.ols('ppvt_raw ~ hard + trapped + work + tired', data=m9_agg).fit()
results.summary()
```

```python
results_agg = smf.ols('ppvt_raw ~ aggravation', data=m9_agg).fit()
results_agg.summary()
```

```python
# Machine Learning part
from sklearn.model_selection import train_test_split 
m9_train, m9_test = train_test_split(m9_agg, test_size = 0.2, random_state=0)
```

```python
lm_agg = smf.ols('ppvt_raw ~ aggravation', data=m9_train).fit()
lm_agg.summary()
# filter out nan values
x = m9_train['ppvt_raw'].fillna(0).values
x = x[x!=0]
#lm_agg.predict()
```

```python
prediction = lm_agg.predict(exog=dict(aggravation=x)).values
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(prediction,x)
```
